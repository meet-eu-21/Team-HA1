```python
from sklearn.decomposition import PCA
from sklearn.neighbors import kneighbors_graph

def pca_knn_basic(X):
    
    pca = PCA(n_components=5)
    pca.fit(X.T)
    print(pca.explained_variance_ratio_)
    print(pca.singular_values_)
    
    edge_index = kneighbors_graph(pca.components_.T, 5, mode='connectivity', include_self=True)
    edge_index = edge_index.toarray()
    edge_index = np.argwhere(edge_index > 0).T
    #graph = graph.diff_op.todense()
    #edge_index = np.array(edge_index)
    print(edge_index)
    
    return edge_index
```

```python
def pca_knn_graphtools(X):
    graph = graphtools.api.Graph(X, n_pca=100, knn=5, decay=40, bandwidth=None, bandwidth_scale=1.0, knn_max=5, anisotropy=0, distance='euclidean', thresh=0.0001, kernel_symm='+', theta=None,precomputed=None, beta=1, sample_idx=None, adaptive_k=None, n_landmark=None, n_svd=100, n_jobs=-1, verbose=False, random_state=None, graphtype='auto', use_pygsp=False, initialize=True)
    #print(graph.data)
    #print(graph.data_pca.components_)
    #print(graph.data_nu)
    #edge_index = graph.diff_op.todense()
    edge_index = graph.K.todense()
    edge_index = np.argwhere(edge_index > 0).T
    #graph = graph.diff_op.todense()
    #edge_index = np.array(edge_index)
    print(edge_index)
    
    return edge_index
```

```python
import os,sys,pickle,time,random,glob
sys.path.append(os.getcwd())
import numpy as np
import pandas as pd

from typing import List
import copy
import os.path as osp
import torch
import torch.utils.data
#from torch_sparse import SparseTensor, cat
from torch_geometric.data import Data, Dataset
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, GATConv
```

```python
parameters = {
    "learning_rate": 0.0001,
    "weight_decay": 0.05,
    "epoch_num": 100,
    "batch_size": 256,
    "number_attention_heads": 8,
    "dropout": 0.4,
    "number_hidden_units": 8,
    "number_node_features": 9,
    "number_classes": 2
}
```

```python
class GAT(torch.nn.Module):
    def __init__(self, parameters):
        super(GAT, self).__init__()
        self.gat1 = GATConv(parameters["number_node_features"], out_channels=parameters["number_hidden_units"],
                            heads=parameters["number_attention_heads"], concat=True, negative_slope=0.001,
                            dropout=parameters["dropout"], bias=True)
        self.gat2 = GATConv(parameters["number_hidden_units"] * parameters["number_attention_heads"], parameters["number_classes"], #n_nodes is missing
                            heads=parameters["number_attention_heads"], concat=False, negative_slope=0.001,
                            dropout=parameters["dropout"], bias=True)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        #x = self.gat1(x.float(), edge_index.float())
        #x = self.gat1(x.double(), edge_index.double())
        x = self.gat1(x, edge_index)
        x = F.elu(x)
        x = self.gat2(x, edge_index)
        return F.log_softmax(x, dim=1)
```
