```python
def get_annotations_dict(binsize, file_CTCF, file_RAD21, file_SMC3, chr_list=None):
    if chr_list is not None:
        chromosomes = chr_list
    else:
        chromosomes = list(file_CTCF.chroms().keys())
        
    dict_values = []
    
    for chromosome in chromosomes:
        chr_len = file_CTCF.chroms()[chromosome]
        gbins = [x for x in range(0, chr_len, binsize)]
        
        annot_CTCF = [np.sum(file_CTCF.values(chromosome, gbin, (gbin+99999))) for gbin in gbins[:-1]]
        annot_RAD21 = [np.sum(file_RAD21.values(chromosome, gbin, (gbin+99999))) for gbin in gbins[:-1]]
        annot_SMC3 = [np.sum(file_SMC3.values(chromosome, gbin, (gbin+99999))) for gbin in gbins[:-1]]

        positions = [(gbin, (gbin+99999)) for gbin in gbins[:-1]]
        values = [(annot_CTCF[i], annot_RAD21[i], annot_SMC3[i]) for i in range(len(annot_CTCF))]
        #dict_one_chr = dict(zip(positions, values))
        dict_values.append(dict(zip(positions, values)))
    
    dictionary = dict(zip(chromosomes, dict_values))
    return dictionary
```
